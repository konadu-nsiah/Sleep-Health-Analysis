```python
mu=0
sigma= 1
N= 1000
X =np.random.normal(mu,sigma,N)
# Population distribution
x_values = np.arange(-10, 10, 0.01)
y_values=norm.pdf(x_values)
# Sample histogram with population distribution
counts, bins,ignored=plt.hist(X, 60, density=True, color="Purple",label="Sampling distribution")
plt.plot(x_values, y_values, color='y', linewidth=2.5, label='population distribution')
plt.title("Showing the normal distribution from customrs' income")
plt.ylabel("Probability")
plt.legend()
plt.show()
```
